5, run python codes to retrive pts coordiates from saved txt file



```python
import pandas as pd
import re
import sys
import os
curr_path = os.getcwd() + '\\'


file_path = curr_path + '\\nx_export_pt\\' + 'test.txt'

# Initialize an empty list to store the coordinates
coordinates = []

# Open the file and read line by line
with open(file_path, 'r') as file:
    for line in file:
        # Search for lines containing "Coordinates [mm]"
        if "Coordinates [mm]" in line:
            # Extract the coordinate values using regex
            match = re.findall(r"[-+]?\d*\.\d+|\d+", line)
            if match and len(match) == 3:
                x, y, z = map(float, match)  # Convert strings to floats
                coordinates.append([x, y, z])

# Convert the list of coordinates to a Pandas DataFrame
df = pd.DataFrame(coordinates, columns=['x', 'y', 'z'])

# Output the DataFrame to a CSV file
output_path = 'test_xyz.txt'
df.to_csv(curr_path + '\\nx_export_pt\\' + 'test_xyz.txt', index=False, header=None)

print(f"Coordinates saved to {output_path}")
```
